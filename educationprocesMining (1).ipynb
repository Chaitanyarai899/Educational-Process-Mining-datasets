```python
Good = df.loc[(df['TOTAL\n(100 points)'] >= 70.0)]
print(len(Good))

Fair = df.loc[(df['TOTAL\n(100 points)'] >= 51.0) & (df['TOTAL\n(100 points)'] <= 69.0)]
print(len(Fair))

Weak = df.loc[(df['TOTAL\n(100 points)'] <= 50.0)]
print(len(Weak))


# Assigning labels to the categories
df.loc[df['TOTAL\n(100 points)'] >= 70.0, 'label'] = 0
df.loc[(df['TOTAL\n(100 points)'] >= 51.0) & (df['TOTAL\n(100 points)'] <= 69.0), 'label'] = 1
df.loc[df['TOTAL\n(100 points)'] <= 50.0, 'label'] = 2

```

```python
df.drop(columns=[
    'Student ID',
],inplace=True)
```

```python
X = df.drop(['label'],axis=1).values
Y = df['label'].values
```

```python
df.head()
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
##Feature Scaling
# Standardization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x_train)
x_train_sc=sc.transform(x_train)
x_test_sc=sc.transform(x_test)
# Normaization
from sklearn.preprocessing import MinMaxScaler
mms= MinMaxScaler()
mms.fit(x_train)
x_train_mms=mms.transform(x_train)
x_test_mms=mms.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
l=LogisticRegression()
l.fit(x_train,y_train)
```

```python
train_pred = l.predict(x_train)
test_pred = l.predict(x_test)
from sklearn.metrics import accuracy_score
train_accur = accuracy_score(y_train,train_pred)
test_accur = accuracy_score(y_test,test_pred)
```

```python
# feature extraction

from sklearn.decomposition import PCA

pca = PCA(n_components = None)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
     
```

```python
# modelling

from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score

permodel = Perceptron()

# Train the model
permodel.fit(x_train, y_train)

y_pred = permodel.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# evaluating the model
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print("RMSE Score :", np.sqrt(mse))

# r2 score
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R2 Score :", r2)
```

```python
# modelling

from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# evaluating the model
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print("RMSE Score :", np.sqrt(mse))

# r2 score
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R2 Score :", r2)
```

```python
# modelling

from sklearn.linear_model import LinearRegression

lmodel = LinearRegression()
lmodel.fit(x_train, y_train)

y_pred =lmodel.predict(x_test)

# evaluating the model
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print("RMSE Score :", np.sqrt(mse))

# r2 score
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R2 Score :", r2)
     
```

```python

# modelling

from sklearn.svm import SVR

svrmodel = SVR()
svrmodel.fit(x_train, y_train)

y_pred = svrmodel.predict(x_test)

# evaluating the model
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print("RMSE Score :", np.sqrt(mse))

# r2 score
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R2 Score :", r2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import r2_score, mean_absolute_error

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
adamodel = AdaBoostClassifier(n_estimators=100, learning_rate=1)

adamodel.fit(x_train, y_train)
y_pred = adamodel.predict(x_test)

print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))

```
